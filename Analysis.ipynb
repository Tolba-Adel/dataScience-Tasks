```python
import pandas as pd
import os
```

# ***Task #1***

**merging all csv files into one**

```python
merged_df=pd.DataFrame()

files=[file for file in os.listdir("/content") if file.endswith(".csv")]
for f in files:
  df=pd.read_csv("/content/"+f)
  merged_df=pd.concat([merged_df,df])

merged_df.to_csv("all_data.csv",index=False)
```

```python
all_data=pd.read_csv("/content/all_data.csv")
all_data
```

# ***Task #2***

**find NaN rows**

```python
nan_rows=all_data[all_data.isna().any(axis=1)] #to show rows with nan values
nan_rows
```

**drop NaN rows**

```python
all_data=all_data.dropna(how='all') #drop those nan rows
all_data
```

**find and**

```python
temp=all_data[all_data['Order Date'].str[0:2]=='Or']
temp
```

**delete the 'Or' in the Month column**

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

**adding a month column**

```python
all_data['Month']=all_data['Order Date'].str.slice(0,2)
all_data['Month']=all_data['Month'].astype('int32')
all_data
```

# ***Task #3***

**convert 'Price Each' and 'Quantity Ordered' columns to float and int respectively**

```python
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data
```

**adding a 'Sales' column**

```python
all_data['Sales']=all_data['Price Each']*all_data['Quantity Ordered']
all_data
```

# ***Question 01:*** *what was the best month for sales ? and how much was earned in that month ?*

```python
months_sales=all_data.groupby('Month').sum()
months_sales
```

***visulizing***

```python
import matplotlib.pyplot as plt

months=range(1,13)

plt.bar(months,months_sales['Sales'])
plt.xticks(months)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

# ***Question 2:*** *what city sold the most products ?*

**add a 'City' column**

```python
all_data['City']=all_data['Purchase Address'].str.split(',').str[1].str.strip()+' '+all_data['Purchase Address'].str.split(',').str[2].str.strip().str.split(' ').str[0]
all_data
```

**calculate the sales sum**

```python
cities_sales=all_data.groupby('City').sum()
cities_sales
```

**plot the results**

```python
cities=[city for city,df in all_data.groupby('City')] #to get the citie in the same order as the groupby('City').sum()

plt.bar(cities,cities_sales['Sales'])
plt.xticks(cities,rotation='vertical')
plt.xlabel('Cities')
plt.ylabel('Sales')
plt.show()
```

# ***Question 3:*** *what time should we display ads to maximize the likelihood of customer's buying a product ?*

**convert the 'Order Date' column's type to a date object**

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

**add an 'Hour' column**

```python
all_data['Hour']=all_data['Order Date'].dt.hour
all_data['Minute']=all_data['Order Date'].dt.minute
all_data
```

**sum the sales per hour**

```python
hours=[hour for hour,df in all_data.groupby('Hour')]

plt.plot(hours,all_data.groupby(['Hour']).count())
plt.grid()
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('Amount of Orders')
plt.show()
```

# ***Question 4:*** *what products are often sold together ?*

**grouping up prodcuts that were sold together**

```python
dup_df=all_data[all_data['Order ID'].duplicated(keep=False)] #get the orders with more then 1 product
dup_df['All Products']=dup_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x)) #group up all the prodcuts of each order
dup_df=dup_df[['Order ID','All Products']].drop_duplicates()
dup_df
```

**counting the pair of prodcuts**

```python
from itertools import combinations
from collections import Counter

count=Counter()
for row in dup_df['All Products']:
  row_list=row.split(',') #convert each row into a list for the 'combinations' method
  count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
  print({key: value})
```

# ***Question 5:*** *which prodcuts sold the most ?*

**grouping up the prodcuts and suming their quantity ordered**

```python
groupby_products=all_data.groupby('Product')
qte_ordered=groupby_products.sum()['Quantity Ordered']
groupby_products
qte_ordered
```

**ploting the results**

```python
products=[product for product,df in groupby_products]

plt.bar(products,qte_ordered)
plt.xticks(products,rotation='vertical')
plt.xlabel('Products')
plt.ylabel('Quantity Ordered')
plt.show()
```

**overlay the graph with the products's prices**

```python
prices=all_data.groupby('Product').mean()['Price Each']

fig,ax1=plt.subplots()
ax2=ax1.twinx()

ax1.bar(products,qte_ordered,color='c')
ax2.plot(products,prices,color='r')

ax1.set_xticklabels(products,rotation='vertical')
ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered',color='c')
ax2.set_ylabel('Price',color='r')

fig.show()
```
